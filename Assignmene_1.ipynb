```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im1 = cv.imread ('emma_gray.jpg',cv.IMREAD_GRAYSCALE )

t = 255 - np.arange(0,256,1).astype(np.uint8)
im2 = t[im1]

plt.plot(t)

cv.namedWindow('Display', cv.WINDOW_NORMAL)
cv.imshow('Display',im1)
cv.waitKey()

cv.imshow('Display',im2)
cv.waitKey()
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im1 = cv.imread ('emma_gray.jpg',cv.IMREAD_GRAYSCALE )

#c =np.array([50,50],[50,100],[150,255],[150,150])

t1 = np.arange(0,50,1).astype(np.uint8)
t2 = np.arange(50,1).astype(np.uint8)
t3 = 22.5 + 1.55*np.arange(50,150,1).astype(np.uint8)
t4 = np.arange(150,1).astype(np.uint8)
t5 = np.arange(150,256,1).astype(np.uint8)

t = np.concatenate((t1,t2,t3,t4,t5),axis=None).astype(np.uint8)
im2 = t[im1]

plt.plot(t)

cv.namedWindow('Display', cv.WINDOW_NORMAL)
cv.imshow('Display',im1)
cv.waitKey()

cv.imshow('Display',im2)
cv.waitKey()
cv.destroyAllWindows()
```
