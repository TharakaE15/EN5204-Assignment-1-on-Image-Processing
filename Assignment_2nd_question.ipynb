```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt


im = cv.imread ('brain_proton_density_slice(1).png',cv.IMREAD_GRAYSCALE )

im_brighter = im + 100
im_brighter_cv = cv.addWeighted(im,1.0,50,1.0,0)

fig,ax = plt.subplots(1,3,figsize =(8,16))
ax[0].imshow(im,cmap='gray',vmin =0,vmax=255)
ax[0].xaxis.tick_top()

ax[1].imshow(im_brighter,cmap='gray',vmin =0,vmax=255)
ax[1].xaxis.tick_top()

ax[2].imshow(im_brighter_cv,cmap='gray',vmin =0,vmax=255)
ax[2].xaxis.tick_top()
   
plt.show()



```
